```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import numpy as np
from numpy import pi
import scipy.stats as stats
from pylab import *
```

```python
# Lee un archivo de texto con la función Pandas "read_csv", con elementos separados por mas de 
# un espacio, brincándose 4 renglones del inicio (encabezados)
df0 = pd.read_csv('datos_2__FFT.txt', skiprows=0, sep='\s+')
```

```python
df0.head()
```

```python
# Dar estructura de datos (DataFrame)
df = pd.DataFrame(df0)
```

```python
# Ver los tipos de datos que Pandas ha reconocido al leer
df.dtypes
```

```python
# Gráfica de y(t)
plt.figure(); df.y.plot(); plt.legend(loc='best')
plt.title("y(t)")
plt.ylabel("y(t)")
plt.xlabel("tiempo")
plt.grid(True)
plt.show()
```

```python
N = 301
T =  1
y = df['y']
yf = fft(y)
xf = fftfreq(N, T)
xf = fftshift(xf)
yplot = fftshift(yf)
import matplotlib.pyplot as plt
plt.plot(xf, 1.0/N * np.abs(yplot))
plt.xlim(0,0.07)
plt.title('Transformada rápida de Fourier')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.grid()
fig=plt.gcf()
fig.set_size_inches(7,7)

plt.show()
```
